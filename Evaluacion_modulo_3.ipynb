```python
# union de datos de los dos archivos por medio del la funcion merge. El left es porque quiero conservar todos los datos, incluso nulo para su posterior limpieza

union_df_df1 = pd.merge(df, df1, on="Loyalty Number", how="left")
union_df_df1
```

2. Limpieza de datos

```python
mis_librerias_1.revisar_nulos(union_df_df1)
```

```python
union_df_df1.head()
```

```python
# aunque hayan 1864 filas duplicadas, he decidido no eliminarlas, ya que los datos no son exactos en todas las columnas. Solo lo son en loyalty number, year, month, y en las otras columnas los datos sí que cambian.Por lo que considero que se puede estructurar con otras funciones como por ejemplo la media en esas columnas. 
```

```python
# en cuanto al porcentaje de nulo en salary, cancelacion_year, cancelacion_month, no los eliminaré.los imputaré con la media en la columna salary.
```

```python
union_df_df1['Salary'] = union_df_df1['Salary'].fillna(union_df_df1['Salary'].median())
```

```python
union_df_df1['Salary'].isnull().sum()
```

```python
# aquí no quise eliminar los nulos porque probablemente son clientes activos que no han cancelado, así que cree una columna para clientes activos 
union_df_df1['Is_Active'] = union_df_df1['Cancellation Year'].isna()
```

```python
union_df_df1.head()
```

```python
union_df_df1.dtypes
```

```python
# aquí preferí cambiar a int el tipo de datos de esas dos columnas, aunque hayan muchos nulos, pero es indispensable para los que no tiene nulos convertirlos a entero y no float. 
union_df_df1['Cancellation Year'] = union_df_df1['Cancellation Year'].astype('Int64')
union_df_df1['Cancellation Month'] = union_df_df1['Cancellation Month'].astype('Int64')
```

_____________ Visualización_________

```python
#  ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?
vuelos_mes= union_df_df1.groupby(['Year', 'Month'])['Flights Booked'].sum().reset_index()
vuelos_mes
```

```python
import calendar

# Lista con nombres de meses en español (puedes cambiar a inglés si quieres)
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
         'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
```

```python
import matplotlib.pyplot as plt

# Agrupar vuelos reservados por mes sumando todos los años
vuelos_mes = union_df_df1.groupby('Month')['Flights Booked'].sum()


plt.figure(figsize=(12,6))
plt.bar(vuelos_mes.index, vuelos_mes.values, color='skyblue')

plt.title('Cantidad total de vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Total vuelos reservados')

plt.xticks(ticks=range(1,13), labels=meses, rotation=45)
plt.tight_layout()
plt.show()
```

# en esta grafica se puede observar que el mes de julio es el que mayor cantidad de vuelos presnta.

```python
# ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente

sns.scatterplot(x = "Distance", 
                y = "Points Accumulated", 
                data = df, 
                size = 10, 
                marker= "v", 
                color = "orange")


plt.xlabel("Distancia")
plt.ylabel("Puntos acumulados");
```

# se observa una allta relacion entre estas dos variables. 

```python

# Contar clientes por provincia
clientes_por_provincia = union_df_df1['Province'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
clientes_por_provincia.plot(kind='bar', color='teal')

plt.title('Distribución de clientes por provincia')
plt.xlabel('Provincia')
plt.ylabel('Número de clientes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# se observa que la provincia de ontario tiene mayor cantidad de clientes.

```python
# ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

salario_promedio_educacion = union_df_df1.groupby('Education')['Salary'].mean().reset_index().sort_values(by='Salary', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=salario_promedio_educacion, x='Education', y='Salary', color='violet')    
```

# la categoria Doctor en la columna educación muestra un mayor salario, en comparacion de las otras categorias

```python
# Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?
clientes_por_tarjeta = union_df_df1.groupby('Loyalty Card').size().reset_index(name='Total Clientes').sort_values(by='Total Clientes', ascending=False)

print(clientes_por_tarjeta)
```

```python

plt.figure(figsize=(8,5))
sns.barplot(data=clientes_por_tarjeta, x='Loyalty Card', y='Total Clientes', color='lightgreen')
plt.title('Clientes por tipo de tarjeta de fidelidad')
plt.xlabel('Tipo de Tarjeta')
plt.ylabel('Número de Clientes')
plt.tight_layout()
plt.show()
```

```python
# ¿Cómo se distribuyen los clientes según su estado civil y género

distribucion = union_df_df1.groupby(['Marital Status', 'Gender']).size().reset_index(name='Total Clientes').sort_values(by='Total Clientes', ascending=False)
print(distribucion)
```

```python

plt.figure(figsize=(10,6))
sns.barplot(data=distribucion, x='Marital Status', y='Total Clientes', hue='Gender', palette='pastel')

plt.title('Distribución de clientes por estado civil y género')
plt.xlabel('Estado Civil')
plt.ylabel('Número de Clientes')
plt.tight_layout()
plt.show()
```
